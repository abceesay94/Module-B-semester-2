```python
import pandas as pd

df = pd.read_csv("EdStatsData.csv (2).zip", compression="zip")
df.head()
```

```python
df.info()
df.describe()
```

The dataset contains educational indicators collected through multiple countries over multiple years from the World Bank. Each row represents a country and indicator combination, while the columns representing the metadata represent the yearly values. Most of the year columns are numerical while the country and indicator are categorical. This dataset is large and spans across multiple decades.

```python
df.duplicated().sum()
```

No duplicated rows were found showing that every row had a unique country-indicator combination.  

```python
df.duplicated(subset=["Country Name", "Indicator Name"]).sum()
```

Each country-indicator should have appear only once. I double checked and there are no duplicate combinations.

```python
df.isnull().sum()
```

There are many missing values across year columns. Some years haven't even happened yet as demonstrated with 2085 - 2100. 
Also, not all countries report each educational indicator every year. Removing all the missing values would significantly reduce the dataset. I figured it would make more sense to focus on a specific indicator and time range to minimize the impact of the missing data.

```python
indicator_name = "Adjusted net enrolment rate, primary, both sexes (%)"

primary_2000 = df[df["Indicator Name"] == indicator_name]["2000"]

primary_2000.mean()

```

```python
indicator_name = "Adjusted net enrolment rate, primary, both sexes (%)"

primary_2000 = df[df["Indicator Name"] == indicator_name]["2000"]

primary_2000.median()

```

```python
indicator_name = "Adjusted net enrolment rate, primary, both sexes (%)"

primary_2000 = df[df["Indicator Name"] == indicator_name]["2000"]

primary_2000.mode()

```

For the indicator "Adjusted net enrolment rate, primary, both sex (%)" in the year 2000, the mean enrollment rate across all the countries was 86.06% while the median was 94.03%. The median is higher than the mean, which suggests that many countries had high enrollment rates, while a smaller number of countries with much lower rates brought the average down. There was no clear mode because each country had a different value. Overall, this shows that enrollment levels varied across countries and that disparities in primary school access still existed at the time.
